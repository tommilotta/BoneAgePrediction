```python
mask_path = "/content/drive/MyDrive/Colab Notebooks/Human Data Analytics/HDA_Project/boneage-training-dataset-masks"
for i in range(1377, 1605):
  im = cv2.imread(mask_path + '/' + str(i) + '.png', 0)
  if im is None:
    continue
  (_, bn) = cv2.threshold(im, 127, 255, cv2.THRESH_BINARY)
  # plt.imshow(bn)
  cv2.imwrite(mask_path + '/' + str(i) + '.png', bn)
```
